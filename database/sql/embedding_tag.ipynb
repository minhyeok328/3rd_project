## Tag to Query

태그를 한 문장의 쿼리로 변환

```python
prompt_tag_to_query = """당신은 식당 검색/추천 시스템을 위한 태그 의미 확장기다.

입력으로 식당의 특징을 나타내는 짧은 태그가 주어진다.
출력은 해당 태그를 임베딩 벡터 생성에 적합한 자연어 문장이다.

목표:
- 짧은 태그를 사용자가 식당을 검색할 때 떠올릴 법한 의미 문장으로 변환한다.
- 식당의 특성, 방문 목적, 이용 상황, 분위기, 운영 방식이 드러나게 쓴다.
- 태그를 단순 반복하지 말고, 실제 검색 의도에 맞게 의미를 풀어쓴다.

규칙:
- 한국어로 작성한다.
- 1~2문장으로 간결하게 작성한다.
- 첫 문장은 태그의 핵심 의미를 식당 특성 중심으로 설명한다.
- 필요하면 두 번째 문장에서 방문 맥락이나 이용 상황을 덧붙인다.
- 광고성 표현, 과장, 주관적 칭찬은 넣지 않는다.
- 태그에 없는 메뉴, 가격, 지역, 업종 정보는 임의로 추가하지 않는다.
- 의미는 풍부하되 불필요하게 장황하지 않게 작성한다.

출력 형식:
- 설명 없이 최종 문장만 하나의 문자열로 출력한다.
- 따옴표 없이 순수 텍스트만 출력한다."""
```

```python
import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def analyze_tag(instr: str) -> str:
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role": "developer",
                "content": prompt_tag_to_query,
            },
            {
                "role": "user",
                "content": instr,
            },
        ],
        temperature=0.3,
        max_completion_tokens=200,
    )

    return response.choices[0].message.content.strip()
```

```python
import time

tag_list = tag_df["name"].unique()
results = []

ind = 0
for tag in tag_list:
    print(f"{ind}번째 작업중")
    results.append(analyze_tag(tag))
    time.sleep(1)
    ind += 1

print(f"{len(tag_list)} : {len(results)}")
print(results[:3])
```

```python
mapping = dict(zip(tag_list, results))
tag_df["description"] = tag_df["name"].map(mapping)
tag_df.head()
```

## Embedding 생성

변환한 쿼리를 embedding 벡터로 변환해서 저장

```python
import base64
import numpy as np
from openai import OpenAI

client = OpenAI(api_key=OPENAI_API_KEY)

def embedding(instr:str) -> str:
    response = client.embeddings.create(
        model='text-embedding-3-small',
        input=[instr, ]
    )
    embedding_vector = response.data[0].embedding

    vec = np.array(embedding_vector, dtype=np.float32)
    blob = vec.tobytes()
    encoded = base64.b64encode(blob).decode('utf-8')

    return encoded
```

```python
import time

description_list = tag_df["description"].unique()
results = []

ind = 0
for description in description_list:
    print(f"{ind}번째 작업중")
    results.append(embedding(description))
    time.sleep(1)
    ind += 1

print(f"{len(description_list)} : {len(results)}")
print(results[:3])
```

```python
mapping = dict(zip(description_list, results))
tag_df["embedding"] = tag_df["description"].map(mapping)
tag_df.head()
```
