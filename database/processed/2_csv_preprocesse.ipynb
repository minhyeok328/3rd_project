## 데이터 라벨링

```python
# menu_df의 price 숫자로 바꾸기
import re
def parse_price(instr):
    if pd.isna(instr):
        return ""

    digits = re.findall(r"\d+", instr)
    return "".join(digits)

menu_df["price"] = menu_df["price"].apply(parse_price)
menu_df.head()
```

```python
# 데이터 라벨링을 위한 고유값 확인
print(review_df["taste_level"].unique()) # 부족, 보통, 좋음
print(review_df["price_level"].unique()) # 불만, 보통, 만족
print(review_df["service_level"].unique()) # 나쁨, 보통, 좋음
```

```python
taste_labels = ['부족', '보통', '좋음']
price_labels = ['불만', '보통', '만족']
service_labels = ['나쁨', '보통', '좋음']

target_labels = []
def lebelize(instr):
    if pd.isna(instr):
        return ""
    
    if instr not in target_labels:
        return instr

    return target_labels.index(instr)

def parse_score(instr):
    return float(instr[:-1])

target_labels = taste_labels
review_df["taste_level"] = review_df["taste_level"].apply(lebelize)
target_labels = price_labels
review_df["price_level"] = review_df["price_level"].apply(lebelize)
target_labels = service_labels
review_df["service_level"] = review_df["service_level"].apply(lebelize)
review_df["reviewer_score"] = review_df["reviewer_score"].apply(parse_score)
review_df.head()
```
