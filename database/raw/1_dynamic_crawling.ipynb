```python
# 펼쳐야 하는 메뉴, 4개 이상의 리뷰가 담긴 페이지
sample_page = "https://www.diningcode.com/profile.php?rid=hTJ7hTGGdjqu"
```

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import re

def read_restaurant(page_link:str) -> str:
    # 드라이버 열기
    driver = webdriver.Chrome()
    driver.get(page_link)

    # 메뉴 열기
    while True:
        try:
            button = driver.find_element(By.CSS_SELECTOR, ".More__Review__Button")
        except NoSuchElementException:
            break

        button_label = button.text

        if button_label != "메뉴 더보기":
            break

        button.click()

        time.sleep(1)
        
    # 루프 횟수 재계산을 위한 리뷰 개수 확인
    # <span class="review_count">(11명의 평가)</span>
    review_count_span = driver.find_element(By.CSS_SELECTOR, "span.review_count")
    matches = re.search(r"\d+", review_count_span.text)
    review_cnt = int(matches.group())
    left_cnt = max(0, review_cnt-4)

    # 리뷰가 다 보일 때까지 버튼 누르기
    for _ in range(left_cnt // 5 + (0 < left_cnt % 5)):
        button = driver.find_element(By.CSS_SELECTOR, ".More__Review__Button")
        button.click()

        time.sleep(1)

    rstr = driver.page_source

    # 다 쓴 페이지 닫기
    driver.close()

    return rstr
```

```python
# 샘플 페이지 파싱 테스트
# 모든 메뉴와 리뷰가 정상적으로 잘 펼쳐져서 나오는지 확인
page_content = read_restaurant(sample_page)
```

```python
with open("page_sample.txt", "w", encoding="utf-8") as f:
    f.write(page_content)
```

```python
# 찾아야 할 100개의 식당 링크 list로 읽어오기
with open("link_restaurants.txt", "r", encoding="utf-8") as f:
    buff = f.read()

page_links = buff.split("\n")
print(len(page_links))
```

```python
# 식당 링크 list 순회하면서 능동 크롤링
for i, link in enumerate(page_links):
    print(f"{i+1}번째 페이지 파싱 중")
    page_content = read_restaurant(link)

    with open(f"restaurants/restaurant_{i}.txt", "w", encoding="utf-8") as f:
        f.write(page_content)

    print(f"{i+1}번째 페이지 파싱 끝")
```
